```python
import pandas as pd
```

```python
# Criando um dataframe a partir do dataset 
df = pd.DataFrame(dataset)
```

```python
# Importando o mÃ³dulo Scikit Learn
# CountVectorizer esse pacote ira contar quantas vezes cada palavra aparece em um conjunto de dados
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Usando o mÃ©todo CountVectorizer para criar uma matriz de documentos
# O fit_transform ira contar as palavras somente na coluna 'text'
cv = CountVectorizer()
count_matrix = cv.fit_transform(df.text)
```

```python
# Contando o nÃºmero de ocorrÃªncias das principais palavras em nosso dataset
word_count = pd.DataFrame(cv.get_feature_names(), columns=["word"])
word_count["count"] = count_matrix.sum(axis=0).tolist()[0]
word_count = word_count.sort_values("count", ascending=False).reset_index(drop=True)
word_count[:100]
```
